## Running Normality tests - Shapiro-Wilk

```python
from scipy.stats import shapiro

stat, p = shapiro(iris_data['petal.length'])
print('Shapiro-Wilk Test for Petal Length:')
print('Statistics=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
    print('Sample looks Gaussian (fail to reject H0)\n')
else:
    print('Sample does not look Gaussian (reject H0)\n')

stat, p = shapiro(iris_data['petal.width'])
print('Shapiro-Wilk Test for Petal Width:')
print('Statistics=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
    print('Sample looks Gaussian (fail to reject H0)\n')
else:
    print('Sample does not look Gaussian (reject H0)\n')

stat, p = shapiro(iris_data['sepal.length'])   
print('Shapiro-Wilk Test for Sepal Length:')
print('Statistics=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
    print('Sample looks Gaussian (fail to reject H0)\n')
else:
    print('Sample does not look Gaussian (reject H0)\n')

stat, p = shapiro(iris_data['sepal.width'])
print('Shapiro-Wilk Test for Sepal Width:')
print('Statistics=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
    print('Sample looks Gaussian (fail to reject H0)\n')
else:
    print('Sample does not look Gaussian (reject H0)\n')
```

## Running Normality tests - Kolmogorov-Smirnof

```python
from scipy.stats import kstest

stat, p = kstest(iris_data['petal.length'], 'norm')
print('Kolmogorov-Smirnof Test for Petal Length:')
print('Statistics=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
    print('Sample looks Gaussian (fail to reject H0)\n')
else:
    print('Sample does not look Gaussian (reject H0)\n')

stat, p = kstest(iris_data['petal.width'], 'norm')
print('Kolmogorov-Smirnof Test for Petal Width:')
print('Statistics=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
    print('Sample looks Gaussian (fail to reject H0)\n')
else:
    print('Sample does not look Gaussian (reject H0)\n')

stat, p = kstest(iris_data['sepal.length'], 'norm')   
print('Kolmogorov-Smirnof Test for Sepal Length:')
print('Statistics=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
    print('Sample looks Gaussian (fail to reject H0)\n')
else:
    print('Sample does not look Gaussian (reject H0)\n')

stat, p = kstest(iris_data['sepal.width'], 'norm')
print('Kolmogorov-Smirnof Test for Sepal Width:')
print('Statistics=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
    print('Sample looks Gaussian (fail to reject H0)\n')
else:
    print('Sample does not look Gaussian (reject H0)\n')

```
